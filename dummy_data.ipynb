## Problem 1

```python
import numpy as np
```

```python
np.random.seed(0)
mean = [-3, 0]
cov = [[1,0.8],[0.8,1]]
rand_number = np.random.multivariate_normal(mean, cov, 500)
print("Create 500 random numbers with a two-dimensional normal distribution: {}".format(rand_number.shape))
```

## Problem 2

```python
from matplotlib import pyplot as plt
```

```python
x, y = rand_number.T
plt.scatter(x, y)
plt.show()
```

## Problem 3

```python
fig = plt.figure()
fig.subplots_adjust(hspace=1)
ax1 = fig.add_subplot(2,1,1)
ax1.set_xlim(-5,2)
ax1.title.set_text("histogram of x1")
ax1.set_xlabel("x1")
ax1.set_ylabel("frequency")
ax1.hist(x)

ax2 = fig.add_subplot(2,1,2)
ax2.set_xlim(-5,2)
ax2.title.set_text("histogram of x2")
ax2.set_xlabel("x2")
ax2.set_ylabel("frequency")
ax2.hist(x)

plt.show()
```

## Problem 4

```python
np.random.seed(0)
mean = [0, -3]
cov = [[1,0.8],[0.8,1]]
rand_number_new = np.random.multivariate_normal(mean, cov, 500)
print("Create 500 random numbers with a two-dimensional normal distribution: {}".format(rand_number_new.shape))
```

```python
x, y = rand_number.T
plt.scatter(x, y, color="b")
x, y = rand_number_new.T
plt.scatter(x, y, color="r")
plt.title("scatter")
plt.xlabel("x1")
plt.ylabel("x2")
plt.legend(['0','1'])
plt.show()
```

## Problem 5

```python
random_number_concat = np.concatenate((rand_number, rand_number_new), axis=0)
print("Combine the ndarrays created in Problems 1 and 4 intoa single ndarray of (1000, 2): {}".format(random_number_concat.shape))
```

```python
random_number_vstack = np.vstack((rand_number, rand_number_new))
print("Combine the ndarrays created in Problems 1 and 4 intoa single ndarray of (1000, 2): {}".format(random_number_vstack.shape))
```
